```python
#Creating a Function to plot categorical variable
def chart_func(col,df):
    dist_data = pd.concat([df[df.TARGET == 1].groupby(col)[col].count()
                     ,df[df.TARGET == 0].groupby(col)[col].count()],axis=1)
    col1,col2 = (col+'_1'),(col+'_0')
    dist_data.columns = [(col1),(col2)]
    
    dist_data_1 = dist_data[col1]/(dist_data[col1]+dist_data[col2])
    
    dist_data_0 = dist_data[col2]/(dist_data[col1]+dist_data[col2])
    
    x = np.arange(len(dist_data))
    width = 0.35

    mpl.style.use('tableau-colorblind10')
    fig, ax = plt.subplots(figsize = (12, 6))
    ax.bar(x - width/2, dist_data_1, width, label='1',color='red')
    ax.bar(x + width/2, dist_data_0, width, label='0',color='green')
    
    ax.set_ylabel('Relative %age')
    ax.set_xlabel(col)
    ax.set_title('TARGET by '+col)
    ax.set_xticks(x)
    ax.set_xticklabels(dist_data.index)
    ax.legend()
    
    fig.tight_layout()
    plt.show()
```

```python
def kde_func(col,df):
    
    mpl.style.use('tableau-colorblind10')
    plt.figure(figsize = (12, 6))
    sns.kdeplot(df.loc[df.TARGET==0][col], label='0',color='green')
    sns.kdeplot(df.loc[df.TARGET==1][col], label='1',color='red')
    
    plt.xlabel(col); 
    plt.ylabel('Density'); 
    plt.title('%s Distribution' % col)
    plt.legend()
    plt.show()
```

```python
#this function aggregates the numerical features of the table using, count, mean, max, min and sum
def numeric_agg(df,group_col,name):
    
    for col in df:
        if col!=group_col and 'SK_ID' in col:
            df=df.drop(columns=col)
    
    var_id = df[group_col]
    df=df.select_dtypes('number')
    df[group_col]=var_id
    
    df_agg = df.groupby(group_col).agg(['count','mean','max','min','sum']).reset_index()
    
    cols = [group_col]
    
    for col in df_agg.columns.levels[0]:
        if col!=group_col:
            for s in df_agg.columns.levels[1][:-1]:
                cols.append('%s_%s_%s' % (name,col,s))
    
    df_agg.columns = cols
    return df_agg
```

```python
#this function aggregates the categorical features of the table using, count and mean
def categorical_agg(df,group_col,name):
    
    categorical = pd.get_dummies(df.select_dtypes('object'))
    categorical[group_col] = df[group_col]
    
    categorical = categorical.groupby(group_col).agg(['sum','mean'])
    
    cols = []
    
    for col in categorical.columns.levels[0]:
        for s in ['count','count_norm']:
            cols.append('%s_%s_%s' % (name,col,s))
        
    categorical.columns = cols
    return categorical.reset_index()
```

```python
#Creating a function generate a dataframe showing %of missing values, sorted in descending value
def generate_missingdf(df):
    miss = df.isnull().sum()
    miss_per = 100*miss/len(df)
    
    miss_df = pd.concat([miss,miss_per],axis=1)
    miss_df.columns=['Values','Values %']
    
    miss_df=miss_df.sort_values('Values %',ascending=False)
    
    return miss_df
```
